```python
mdf=pd.merge(deliveries,match,left_on='match_id',right_on='id')
mdf
```

```python
mdf.info()
```

```python
cd=deliveries.copy()
cd.head(10)
```

```python
# creating batsman dataframe

# run scored
bat_df=mdf.groupby(by='batsman')['batsman_runs'].sum().reset_index() # run scored

# ball faced
bat_df['Ball']=mdf.groupby(by='batsman')['ball'].count().reset_index()['ball']  

# no. of sixes
bat_df['sixes']=cd.groupby(by='batsman')['batsman_runs'].apply(lambda x:(x==6).sum()).reset_index()['batsman_runs']

# no.of fours
bat_df['fours']=cd.groupby(by='batsman')['batsman_runs'].apply(lambda x:(x==4).sum()).reset_index()['batsman_runs']

# changing my single column name
bat_df.rename(columns={'batsman_runs':'Run_scored'},inplace=True)

# strikerate
bat_df['strikerate']=bat_df['Run_scored']/bat_df['Ball']

# player of match
bat=bat_df.copy()

PL=match['player_of_match'].value_counts().reset_index()

g=pd.merge(bat,PL,left_on='batsman',right_on='index',how='left')

g=g.drop('index',1)

g.rename(columns={'player_of_match':'count_player_of_match'},inplace=True)

g['count_player_of_match'].fillna(0,inplace=True)

g['count_player_of_match']=g['count_player_of_match'].apply(round)

# highest score in match

HR=mdf.groupby(by=['match_id','batsman'])['batsman_runs'].sum().reset_index()
g['max_score']=HR.groupby(by='batsman')['batsman_runs'].max().reset_index()['batsman_runs']

# to find count of match for each batsman
g['no_of_matches']=cd.groupby(by='batsman')['match_id'].count().reset_index()['match_id']

# to find out batting team
g['team_played']=cd.groupby(by='batsman')['batting_team'].unique().reset_index()['batting_team']

g.head(5)
```

```python
# bowlers data frame

# to find out total no. of overs
bowler_df=cd.groupby(by=['bowler','bowling_team'])['over'].nunique().reset_index()

# to find out total run given by bowler
bowler_df['run_given']=cd.groupby(by=['bowler','bowling_team'])['total_runs'].sum().reset_index()['total_runs']

# find out total extra runs given by bowler
bowler_df['total-extra_run']=cd.groupby(by=['bowler','bowling_team'])['extra_runs'].sum().reset_index()['extra_runs']

# no of wides by bowler
bowler_df['no_of wide']=cd.groupby(by=['bowler','bowling_team'])['wide_runs'].nunique().reset_index()['wide_runs']


# total no ball rus runs
bowler_df['no_ball_run']=cd.groupby(by=['bowler','bowling_team'])['noball_runs'].sum().reset_index()['noball_runs']

# player of match
bal=bowler_df.copy()
h=pd.merge(bal,PL,how='left',left_on='bowler',right_on='index')
h.drop('index',1,inplace=True)
h['player_of_match'].fillna(0,inplace=True)

#h['player_of_match'].apply(lambda x:round(x))

itn=h['player_of_match']
mad=itn.map(round)
h['count_of-pl_mat']=mad
h.drop('player_of_match',1,inplace=True)

# economy
h['economy']=h['run_given']/h['over']

# no of wickets taken

h['wickets']=cd.groupby(by=['bowler','bowling_team'])['player_dismissed'].count().reset_index()['player_dismissed']

h.head(10)
```
